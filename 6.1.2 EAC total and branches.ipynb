```python
def time_series_1_1deg():
    
    plt.figure(figsize(12,8))
    
    t1 = []
    t2 = []
    t3 = []
    t4 = []
    
    for tt in range (1,184,1):
        t1.append(tt)
        
    for tt in range (1,184,10):
        t2.append(tt)
        
    plt.plot(t1,xx1, 'r',label = 'EAC upstream transport 1deg_1',linewidth = 2)
    plt.plot(t1,yy1, 'r--',label ='EAC sum of branches transport 1deg_1',linewidth = 2)
    
    for tt in range (1,175,1):
        t3.append(tt)
        
    for tt in range (1,175,10):
        t4.append(tt)
        
    plt.plot(t3,x1, 'r',label = 'EAC upstream transport 1deg_10',linewidth = 3)
    plt.plot(t3,y1, 'r--',label ='EAC sum of branches transport 1deg_10',linewidth = 3)
    
    plt.legend(loc = 'upper right')
    
                        
    plt.xticks(t4)
    plt.title('Time series of EAC upstream and sum of branches transport')
    plt.xlabel('year')
    plt.ylabel('(Sv)')
    # plt.ylim([-15,-4])
    # plt.ylim([-15,6])
    
    savefigure('1')
```

### Scatter plot and linear regression

```python
def scatter_plot():
    plt.figure(figsize(12,8))
    plt.scatter(x2, y2, label = '025deg', color = 'b')
    
    plt.plot(x2,x2,'r', label = 'x=y')
    
    trend = np.polyfit(x2, y2, 1)
    y = trend[1] + x2*trend[0]
    print(trend[1])
    print(trend[0])
    plt.plot(x2, y, color = 'k', label = 'relationship')
    
    plt.legend(loc = 'upper right')
    plt.title('relationship between EAC upstream trnasport and sum of EAC branches transport')
    plt.xlabel('EAC upstream transport (Sv)')
    plt.ylabel('EAC sum of branches transport (Sv)')
    
    savefigure('2_02') 
```
